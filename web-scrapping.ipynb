```python
html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story: a story of brothers and sisters</b></p>

<p id="story">
Once upon a time there were three little sisters and two little brothers; and their names were
    <a href="http://example.com/elsie" class="sister" id="girl1">Elsie</a>,
    <a href="http://example.com/lacie" class="sister" id="girl2">Lacie</a> and
    <a href="http://example.com/tillie" class="sister" id="girl3">Tillie</a>;
    <a href="http://example.com/tahar" class="brother" id="boy1">Tahar</a>;
    <a href="http://example.com/merzouk" class="brother" id="boy2">Merzouk</a>;
and they lived at the bottom of a well.
</p>

<p id="details">
    <div id="girl1-details" class="sister">
        <p class="name">Elsie</p>
        <p class="age">15<p>
    </div>
    <div id="girl2-dehails" class="sister">
        <p class="name">Lacie</p>
        <p class="age">17<p>
    </div>
    <div id="girl3-details" class="sister">
        <p class="name">Tillie</p>
        <p class="age">25<p>
    </div>
</p>
"""
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_doc)
```

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
soup.p
```

```python
soup.p.string
```

```python
soup.find_all('a')
```

```python
soup.find_all('a', class_='sister')
```

```python
soup.find(id="girl1")
```

```python
soup.select('.sister')
```

```python
soup.select_one('.sister')
```

```python
soup.select('#story .sister')
```

```python
soup.select('#story .brother')
```

```python
soup.select('#details .sister .name')
```

```python
sisters_names = soup.select('#details .sister .name')
for idx, el in enumerate(sisters_names):
    print(f'Name of sister N° {idx+1} is {el.string}')
```

```python
sisters = soup.select('#details .sister')
for idx, el in enumerate(sisters):
    name = el.select_one('.name').get_text().strip()
    age = el.select_one('.age').get_text().strip()
    print(f'Sister N° {idx+1}: name={name}, age={age}')
```
