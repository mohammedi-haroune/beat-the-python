```python
import csv
```

```python
path = "products.csv"
with open(path, "r", newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    products = list(reader)
```

```python
fields = ['name','created_at', 'price', 'advertiser', 'type']

with open("products_mini.csv", "w", newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fields)
    writer.writeheader()
    for product in products:
        writer.writerow({k:v for k,v in product.items() if k in fields})
```

```python
MINIMUM_PRICE = 10000.

products_price_fixed = []

with open("products_mini.csv", newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for product in reader: 
        try:
            price, _, fix_or_negociable = product['price'].split()
        except ValueError:
            print(f'Skipping {product}, cannot decode the price')
            continue

        if float(price) < MINIMUM_PRICE:
            continue
                
        product['fixed_price'] = fix_or_negociable == 'Fixe'
        product['price'] = float(price)

        products_price_fixed.append(product)
```

```python
fields = ['name','created_at', 'price', 'fixed_price', 'advertiser', 'type']

with open("products_fixed_price.csv", "w", newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fields)
    writer.writeheader()
    for product in products_price_fixed:
        writer.writerow({k:v for k,v in product.items() if k in fields})
```
